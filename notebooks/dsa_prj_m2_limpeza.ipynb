#Filtro Inicial

```python
cv = dados_cv[dados_cv.columns.drop('Total')]
cv = cv[cv.columns.drop(cv.columns[1:8])]
cv.head()
```

```python
casos = dados_casos[dados_casos.columns[:-2]]
casos.head()
```

```python
fxetaria = dados_fx_etaria.drop(columns='Total')
fxetaria.head()
```

```python
sexo = dados_sexo.drop(columns='Total')
sexo.head()
```

#Criando função para limpar DataFrames, aplicando o seguinte padrão.

*   Colocando a `Unidade de Federação` como índice e renomeando para `uf`
*   Removendo os números que antecedem o nome dos estados
*   Convertendo as colunas para o tipo `int`



```python
def limpeza(df):
  uf = df.columns[0]
  df = df.rename(columns={uf:'uf'})
  df = df.set_index('uf')
  df.index = df.index.str[3:]
  return df
```

```python
def vazio(df: pd.DataFrame):
  qtd = df.isna().sum().sum()
  qtd = qtd + df.isnull().sum().sum()
  return qtd
```

#Aplicando limpeza para os DataFrames

```python
cv = limpeza(cv)
casos = limpeza(casos)
fxetaria = limpeza(fxetaria)
sexo = limpeza(sexo)
```

#Verificando dados incompletos

```python
display(vazio(cv))
display(vazio(casos))
display(vazio(fxetaria))
display(vazio(sexo))
```
